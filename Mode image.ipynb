```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('Ling_image.jpg')
grayscale = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(30,8))
ax1.set_title('BRG')
ax1.imshow(img)
ax2.set_title('RGB')
ax2.imshow(rgb)
ax3.set_title('GRAY')
ax3.imshow(grayscale, cmap = 'gray')
```

```python
hist, bins = np.histogram(grayscale.flatten(),256, [0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc ='upper left')
plt.show
```

```python
text = "Love"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1.0
font_thickness = 2
text_color = (255, 255, 255)  # White color in BGR
text_position = (50, 150)  # (x, y) coordinates
image_with_text = cv2.putText(img, text, text_position,font, font_scale, text_color, font_thickness )
cv2.imshow('Ling', image_with_text)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
